# Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

### Load training datasets

```python
training_pres_df = pd.read_csv('./Training Datasets/training_pres.csv')
training_pres_df.head()
```

## Vote

### Analysis of the data

```python
race_columns = ['White alone', 'Black or African American alone', 'American Indian and Alaska Native alone', 
                'Asian alone', 'Native Hawaiian and Other Pacific Islander alone', 'Some Other Race alone']
race_sums = training_pres_df[race_columns].sum()

plt.figure(figsize=(10, 7))
plt.pie(race_sums, labels=race_columns, autopct='%1.1f%%')
plt.title('Distribution of Races in People Data')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
plt.hist(training_pres_df['Occupied_percent'], bins=np.arange(0, 1.1, 0.01))
print("Occupied Housing Percentage Mean: ", training_pres_df['Occupied_percent'].mean())
plt.axvline(training_pres_df['Occupied_percent'].mean(), color='r', linestyle='dashed', linewidth=1, label='Mean')
plt.legend()
plt.xlabel('Occupied Housing Percentage')
plt.ylabel('Number of Precincts')
plt.title('Housing Occupancy Distribution')
plt.show()
```

```python
plt.hist(training_pres_df['democrat_percent'], bins=np.arange(0, 1.1, 0.01))
print("Democrat Percent Mean: ", training_pres_df['democrat_percent'].mean())
plt.axvline(training_pres_df['democrat_percent'].mean(), color='r', linestyle='dashed', linewidth=1, label='Mean')
plt.legend()
plt.xlabel('Democrat Percent')
plt.ylabel('Frequency')
plt.title('Histogram of Democrat Percent in 2020 Presidential Election in IL7')
plt.show()
```

```python
columns_of_interest = race_columns + ['Occupied_percent', 'democrat_percent']
correlation_matrix = training_pres_df[columns_of_interest].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='RdBu', vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
columns_of_interest_narrowed = ['White alone', 'Black or African American alone', 'Asian alone', 'Occupied_percent', 'democrat_percent']
correlation_matrix_narrowed = training_pres_df[columns_of_interest_narrowed].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_narrowed, annot=True, cmap='RdBu', vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
plt.hist(training_pres_df['Turnout'], bins=np.arange(0, 1.1, 0.01))
print("Number of registered voters: ", training_pres_df['Registered'].sum())
print("Turnout mean: ",training_pres_df['Turnout'].mean())
plt.axvline(training_pres_df['Turnout'].mean(), color='r', linestyle='dashed', linewidth=1, label='Mean')
plt.legend()
plt.xlabel('Turnout')
plt.ylabel('Frequency')
plt.title('Histogram of Turnout in 2020 Presidential Election in IL7')
plt.show()
```

```python
columns_of_interest_with_turnout = columns_of_interest + ['Turnout']
correlation_matrix_with_turnout = training_pres_df[columns_of_interest_with_turnout].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_with_turnout, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap with Turnout')
plt.show()
```

```python
columns_of_interest_with_turnout_narrowed = columns_of_interest_narrowed + ['Turnout']
correlation_matrix_with_turnout_narrowed = training_pres_df[columns_of_interest_with_turnout_narrowed].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_with_turnout_narrowed, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap with Turnout')
plt.show()
```
